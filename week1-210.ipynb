# 10. Function

```python
def power(x, y):
    return x**y

power(2,10)
```

ต้องมีคำสั่ง def เสมอ และกด tab ต่อด้วย return ข้อนี้เป็ยการแสดงการยกกำลัง

```python
def concat(A, B):
    A = A + " First"
    B = B + " Second"
    C = A + " " + B + " Combined"
    return A,B,C

x, y, z = concat("Pladook","Here")
[x] + [y] + [z]
```

code จะรู้เลยว่า A,B,C ที่ถูก return มา คือตัวแปร [x] + [y] + [z] ที่อยู่ด้านล่าง

### 2. Write a Python program to count the number of even and odd numbers from a series of numbers.

```python
L = [i for i in range(1,10)]


odd = [x for x in L if x % 2 == 1] 
even = len(odd) 

print("Number of even numbers: ", even) 
print("Number of odd numbers : ", len(L) - even) 
```

len เป็นคำสั่งสำหรับใช้วัดความยาวของตัวอักษร ใช้ได้กับทั้งสตริง, ไบต์, ทูเพิล, List, หรือ range

โดยใน code คือการ range ค่า1-10 ขึ้นมา แล้วสร้างตัวแปร odd ขึ้นมาเพื่อเก็บค่า x ตัวที่ mod 2 แล้วมีค่าเท่ากับ 1 ไว้
หลังจากนั้นส้รางตัวแปร even ขึ้นมาเพื่อ len ค่าในตัวแปร odd ที่เราเก็บไว้ และหาจำนวน even โดยเอา len ของ L มาลบ len ของ odd เพื่อให้ได้ len ของ even มาตอบ
