# 2. Sub Plot

```python
plt.figure(1, figsize=(15,3))    # first create 1 figure  #figure ที่เลข1คือการใส่ไอดีให้มัน figsizeคือการกำหนดขนาดกราฟใช้เป็น "นิ้ว"                              # the figsize argument is a tuple of (width, height)
plt.subplot(1,3,1)               # numrows=1, numcols=3, fignum=1 #คือการบอกว่าให้ใน 1 แถวมี 3 รูป แล้วจะพล็อตรูปที่1
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'$f(x) = x$')         # ใส่ r'เพื่อให้เป็นตัวเอียง
plt.plot(y1, color='tan')        # plot the y1 sequence  #สีเส้นกราฟ

plt.subplot(1,3,2)               # numrows=1, numcols=3, fignum=2 #คือการบอกว่าให้ใน 1 แถวมี 3 รูป แล้วจะพล็อตรูปที่2
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'$f(x) = x^2$')       # scientific "TeX" notation for superscript
plt.plot(y2, color='darkorange') # plot the y2 sequence

plt.subplot(1,3,3)               # numrows=1, numcols=3, fignum=3  #คือการบอกว่าให้ใน 1 แถวมี 3 รูป แล้วจะพล็อตรูปที่3
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'$f(x) = x^3$')
plt.plot(y3, color='green')      # plot the y3 sequence

plt.show()
```

# 3. Graph Type

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(1, figsize=(15,6))

# bar chart example ใช้เปรียบเทียบมากน้อย
accounts = ['fabiyamada', 'jedigeiss', 'scipio', 'stoodkev']
steem = [1100, 500, 1200, 2300]

plt.subplot(1,2,1)    #1 row 2 column plotรูปที่1
plt.title('Bar chart')
plt.ylabel('Steem')
plt.bar(accounts, steem) #ชื่อแกนx และแกนy #accountคือการดึงค่า account ที่ตั้งไว้ด้านบนมาใช้

# pie chart example ข้อมูลข้างในต้องเป็น % เท่านั้น
labels = ['fabiyamada', 'jedigeiss', 'scipio', 'stoodkev']
perc = [steem[0]/sum(steem), # calculate percentages from the amount of steem (1100/ทั้งหมด)
        steem[1]/sum(steem), 
        steem[2]/sum(steem), 
        steem[3]/sum(steem)]

explode = (0, 0, 0.1, 0)     # "explode" the 3rd slice ('Scipio') ระยะห่างของแต่ละอัน

plt.subplot(1,2,2)
plt.title('Pie chart')
plt.axis('equal')            # Make sure the pie is drawn as a circle, X and Y axes equally sized #ให้แกนx แกนy ของกราฟมีขนาดเท่ากัน
plt.pie(perc, labels=labels, autopct='%1.2f%%', 
        startangle=90, explode=explode) #startangle(องศาที่ต้องการเริ่ม "เส้นที่อยู่ตรงคำว่า pie chart") #explode การดึงค่าexplodeที่ตั้งไว้ข้างบนมาใช้
plt.show()
```
